```python
import tensorflow as tf
import pandas as pd
import numpy as np
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.layers import Bidirectional

```

```python
df = pd.read_csv('./assets/dataset_sms_spam_v1.csv')
xs = df['Teks']
ys = df['label']
```

```python
OOV_TOKEN = "<OOV>"
MAX_VOCAB = 100
EMBEDDING_DIM = 8
```

```python
tokenizer = Tokenizer(oov_token=OOV_TOKEN)
tokenizer.fit_on_texts(xs)
tokenizer.word_index
```

```python
sequences = tokenizer.texts_to_sequences(xs)
padded_sequences = pad_sequences(sequences, padding='post',maxlen=50)
padded_sequences.shape
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(100,EMBEDDING_DIM,input_length=50),
    Bidirectional(tf.keras.layers.LSTM(56)),
    tf.keras.layers.Dense(54,activation='relu'),
    tf.keras.layers.Dense(3,activation='softmax')
])
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(padded_sequences, ys, epochs=30)
```
